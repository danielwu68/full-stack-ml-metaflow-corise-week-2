```python
from metaflow import Flow
name = 'TFlow5'
run = Flow(name).latest_run
print(run.successful)
```

```python
run.data.results
```

```python
import matplotlib.pyplot as plt

# x = run.data.results[0][0].__dict__
# display(x)

def to_label(row):
    l = row[0]
    return f"{l.__class__}-{l.max_depth}"

results = sorted([*run.data.results], key=to_label)
display(results)

labels = [to_label(r) for r in results]
values = [r[1] for r in results]
# display(labels, values)

plt.plot(labels, values)
plt.xticks(rotation=90)
plt.show()
```
